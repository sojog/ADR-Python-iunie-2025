```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [20, 30, 40, 30, 10]
plt.pie(x=data)
plt.show()
```

```python
data = [20, 30, 20, 30, 10]
plt.pie(x=data,labels=data)
plt.show()
```

```python
data =  [ 30, 20, 10, 30, 10]

percentage = [ f"{el * 100 / sum(data)} %"   for el in data ]

plt.pie(x=data,labels=percentage)
plt.show()
```

```python
data =  np.array([20, 30, 20, 20, 10])

percentage = [ f"{el * 100 / sum(data)} %"     for el in data ]

plt.pie(x=data,labels=percentage)
plt.show()
```
